```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') #membuat data dummy, tips adalah data yang adaa di seaborn
tips
```

```python
tips.head(5)
```

```python
#distplot
sns.distplot(tips['total_bill'])
```

```python
#ingin menampilkan tidak ada garis dan datanya ada 10 buah
sns.distplot(tips['total_bill'],kde=False,bins=10) #10 kotak histogram
```

```python
#joint plot(menggabungkan plot.contoh=plot scatter dan histogram)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#pairplot(data yang )
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='RdPu')
```

```python
#rugplot
sns.rugplot(tips['total_bill'])
```
