```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips) #counterplot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='Reds')
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='RdPu')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='Reds') #data yang gendut berarti angka di y lebih banyak
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='Reds',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, palette='rainbow',hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, palette='rainbow',hue='sex',jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, palette='rainbow',hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
#kombinasi swarm dan violin
sns.violinplot(x='tip',y='day',palette='rainbow',data=tips)
sns.swarmplot(x='tip',y='day',color='black',data=tips,size=3)
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')
```
