```python
import numpy as np
import pandas as pd
```

```python
#membuat tabel
data = {'Bagian':['Adv','Adv','HR','HR'],
        'Nama':['Dian','Nur','Yani','Angel'],
        'Penjualan':[100,200,300,400]}
df=pd.DataFrame(data)
df
```

```python
#concat
data1 = {'A':[1,2,3,4],
         'B':[1,2,3,4],
         'C':[1,2,3,4],
         'D':[1,2,3,4]}
data2 = {'A':[1,2,3,4],
         'B':[1,2,3,4],
         'C':[1,2,3,4],
         'D':[1,2,3,4]}
data3 = {'A':[1,2,3,4],
         'B':[1,2,3,4],
         'C':[1,2,3,4],
         'D':[1,2,3,4]}
df1=pd.DataFrame(data1,index=[0,1,2,3])
df2=pd.DataFrame(data2,index=[4,5,6,7])
df3=pd.DataFrame(data3,index=[8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
tabelbaru=pd.concat([df1,df2,df3])
tabelbaru
```

```python
tabelbaru2=pd.concat([df1,df2,df3],axis=1)
tabelbaru2
```

```python
#merge
left = {'key1':['K0','K0','K1','K1'],
        'key2':['K0','K0','K0','K0'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']}
right = {'key1':['K0','K1','K1','K2'],
        'key2':['K1','K2','K0','K3'],
        'C':['C0','C1','C2','C3'],
        'D':['D0','D1','D2','D3']}
dfleft = pd.DataFrame(left)
dfright = pd.DataFrame(right)
```

```python
dfleft
```

```python
dfright
```

```python
pd.merge(dfleft,dfright,how='outer',on=['key1','key2'])
```

```python
pd.merge(dfleft,dfright,how='inner',on=['key1','key2'])
```

```python
pd.merge(dfleft,dfright,how='right',on=['key1','key2'])
```

```python
pd.merge(dfleft,dfright,how='left',on=['key1','key2'])
```

```python
pd.merge(dfleft,dfright,how='outer',on=['key1','key2'])
```

```python
left = {'A':['A0','A1','A2'],
        'B':['B0','B1','B2']}
right = {'C':['C0','D1','C2'],
         'D':['D0','D1','D2']}
dfleft = pd.DataFrame(left)
dfright=pd.DataFrame(right,index=[3,4,5])
```

```python
dfleft
```

```python
dfright
```

```python
dfleft.join(dfright,how='outer')
```

```python
dfleft.join(dfright,how='right')
```

```python
dfleft.join(dfright,how='left')
```
