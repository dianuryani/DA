```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#tipe2 regregasi
#lmplot
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,palette='coolwarm')
```

```python
#modifikasi scatter (bulat-bulat)
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,palette='coolwarm',markers=['o','v'],scatter_kws={'s':100}) #scatter_kws untuk memperbesar m
```

```python
#memisahkan diagram berdasarkan sex
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,palette='coolwarm',markers=['o','v'],scatter_kws={'s':100},col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,palette='coolwarm',markers=['o','v'],scatter_kws={'s':100},col='day',
           aspect=1,height=8) #aspect mengatut lebar grid, height mengatur tinggi grid
```
