```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
#membuat plot
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot Contoh')
plt.show()
```

```python
#membuat 2 buah plot dalam 1 canva
plt.subplot(1,2,1) #row 1 kolom 2 #plot 1
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
#membuat plot ukuran custon menggunakan figure
fig = plt.figure()

axes = fig.add_axes([1,1,1,1]) #jarak dari kiri,jarak dari bawah, lebar canvas, tinggi canvas

axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh judul')
```

```python
#membuat figure ada 2plot
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X utama')
axes1.set_ylabel('Y utama')
axes1.set_title('plot dalam')

axes2.plot(y,x,'r')
axes2.set_xlabel('X dalam')
axes2.set_ylabel('Y dalam')
axes2.set_title('plot luar')
```

```python
#mengatur hasil jadi agar kualitas foto bagus
fig = plt.figure(figsize=(1000,1000), dpi=100) #fisize=ukuran canvas lebar, tinggi. dpi=makin besar, makin tajam
```

```python
fig,axes = plt.subplots(figsize=(6,4), dpi=100)

#atau
#fig =plt.figure()
#axes = fig.add_axes([1,1,1,1])

axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('plot')

fig.savefig('contoh png', dpi=100)
```

```python
#memberikan legend

#fig = plt.figure()
#axes = fig.add_axes([1,1,1,1])
#ATAU
fig,axes = plt.subplots(figsize=(6,4), dpi=100)

axes.plot(x,x**2,'r',label='Data penjualan')
axes.plot(x,x**3,'y',label='Data pembelian')

axes.legend(loc=0)
```

```python
fig,axes = plt.subplots(figsize=(6,4), dpi=100)

axes.plot(x,x+2,color='red',label='y1',lw=0.5,ls='--')
axes.plot(x,x+3,color='red',label='y2',lw=1,ls='--')
axes.plot(x,x+4,color='red',label='y3',lw=1.5,ls='--')
axes.plot(x,x+5,color='red',label='y4',lw=2,ls='--')
```

```python
fig,axes = plt.subplots(figsize=(6,4), dpi=100)

axes.plot(x,x+2,color='red',label='y1',lw=0.5,ls='-')
axes.plot(x,x+3,color='red',label='y2',lw=1,ls='--')
axes.plot(x,x+4,color='red',label='y3',lw=1.5,ls='-.')
axes.plot(x,x+5,color='red',label='y4',lw=2,ls=':')
```

```python
#mengatur jarak linestyle
fig,axes = plt.subplots(figsize=(6,4), dpi=100)

line, = axes.plot(x,x+2,color='black',label='y1',lw=2,ls='-')
line.set_dashes([10,1,2,2]) #2 awal line style, 2 akhir yaitubaris
#panjang awal ls, jarak ls, panjang baris, jarak baris
```

```python
#variasi linestyle
fig,axes = plt.subplots(figsize=(6,4), dpi=100)

axes.plot(x,x+2,color='blue',label='y1',lw=1,ls='--',marker='d',markersize=5)
axes.plot(x,x+3,color='blue',label='y1',lw=1,ls='--',marker='o',markerfacecolor='yellow',markeredgewidth=1,markeredgecolor='black')
axes.plot(x,x+4,color='blue',label='y1',lw=1,ls='--',marker='s')
```
