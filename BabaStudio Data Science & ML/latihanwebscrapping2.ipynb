```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import math
```

```python
#jalankan webdriver chrome yang telah di download
driver = webdriver.Chrome('chromedriver')
```

```python
#webdriver akan akses web yang akan di scrapping
#webscrapping dari tokopedia
#membuat array kosong untuk menampung info product

products=[]
prices=[]
images=[]

driver.get('https://shopee.co.id/search?keyword=laptop')
```

```python
content = driver.page_source #page souurce dari html page yang diteliti
soup = BeautifulSoup(content) #soup variabel yang menampung kode html dari content agar dapat diakses/diambil isi kontennya nanti
soup #(html website yang telah dikunjungi)
```

```python
for link in soup.find_all('div',class_='col-xs-2-4 shopee-search-item-result__item'): #mencari semua link yg ada di soup bertag div dengan class disamping
    print(link)
```

```python
#mendapat data image
for link in soup.find_all('div',class_='col-xs-2-4 shopee-search-item-result__item'):
    dataimg = link.find('img',class_='_1T9dHf _3XaILN')
    print(dataimg["src"])
```

```python
#mendapat nama/judul laptopnya
for link in soup.find_all('div',class_='col-xs-2-4 shopee-search-item-result__item'):
    name = link.find('div',class_='_1NoI8_ _16BAGk')
    print(name.get_text())
```
