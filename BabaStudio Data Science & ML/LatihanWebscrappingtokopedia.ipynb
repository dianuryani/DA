```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
#jalankan webdriver chrome yang telah di download
driver = webdriver.Chrome('chromedriver')
```

```python
#webdriver akan akses web yang akan di scrapping
#webscrapping dari tokopedia
#membuat array kosong untuk menampung info product

products=[]
prices=[]
images=[]

driver.get('https://www.tokopedia.com/search?st=product&q=samsung')
```

```python

content = driver.page_source #page souurce dari html page yang diteliti
soup = BeautifulSoup(content) #soup variabel yang menampung kode html dari content agar dapat diakses/diambil isi kontennya nanti
soup #(html website yang telah dikunjungi)
```

```python
for link in soup.find_all('div',class_='css-1g20a2m'): #mencari semua link yg ada di soup bertag div dengan class disamping
    print(link)
```

```python
#mendapat data image
for link in soup.find_all('div',class_='css-1g20a2m'):
    dataImg = link.find('img',class_='success fade')
    print(dataImg['src'])
```

```python
#mendapat nama/judul laptopnya
for link in soup.find_all('div',class_='css-1g20a2m'):
    name = link.find('div',class_='css-18c4yhp')
    print(name.get_text())
```

```python
#mendapat price laptopnya
for link in soup.find_all('div',class_='css-1g20a2m'):
    price = link.find('div',class_='css-rhd610')
    print(price.get_text())
```

```python
#mengisi data yang telah diperoleh ke variabel array yang telah disiapkan sebelumnya
for link in soup.find_all('div',class_='css-1g20a2m'):
    name = link.find('div',class_='css-18c4yhp')
    #print(name.get_text())
    dataImg = link.find('div',class_='success fade')
    #print(dataImg['src'])
    price = link.find('div',class_='css-rhd610')
    #print(price.get_text())
    
    products.append(name.get_text())
    images.append(dataImg)
    prices.append(price.get_text())
```

```python
df = pd.DataFrame({'Product Name': products, 'Price': prices, 'Images':images})
df
```

```python
#membuat ke dalam bentuk file csv
df.to_csv('tokopedia.csv',index=False,encoding='utf-8')
```

```python
#membuat ke dalam bentuk file json
df.to_json('tokopedia.json')
```
